# Analisi di Hash Table
In questo notebook si vuole presentare un'analisi relativa alle prestazioni delle **hash table**, strutture dati particolarmente efficaci per implementare dizionari, ovvero insiemi dinamici in grado di supportare le operazioni di *inserimento*, *ricerca* e *cancellazione* di un elemento. 

Una hash table è costituida da un insieme di elementi, ad ognuno dei quali viene associata una chiave. L'accesso ad un determinato elemento viene effettuato attraverso la chiave. Nella seguente analisi si assume che le chiavi sono tra loro distinte.

# 1. Direct Address Table
Una hash table generalizza il concetto di **array**. Un array può infatti essere utilizzato per implementare una tabella in cui la chiave di ogni elemento coincide con la posizione del vettore in cui l'elemento è memorizzato.

Grazie all'**indirizzamento diretto**, tutte le operazioni su un array vengono eseguite in un tempo $O(1)$.
Tuttavia, dal momento che tipicamente la dimensione del vettore $m$ coincide con la cardinalità dell'insieme delle chiavi $\left|K\right|$, tale soluzione risulta poco efficiente da un punto di vista di occupazione di memoria se il numero effettivo di chiavi memorizzate $n$ è inferiore a $\left|K\right|$.

### Insert

#### Codice
```java
public boolean insert(Element element) {
    if(table[element.getKey()] == null){
        table[element.getKey()] = element;
        n = n + 1;
        return true;
    } else {
        return false;
    }
}
```

#### Analisi
L'analisi è stata effettuata considerando i seguenti parametri:
- $m$ dimensione dell'array: $m = 10^6$
- $n$ numero di elementi effettivamente memorizzati: $n \in \left[0, m-10^3\right]$
- *n_step* passo con cui $n$ è incrementato: *n_step* $= 10^3$ 
- *op_iteration* numeri di volte in cui un'operazione è eseguita, per ogni $n$: *op_iteration* $= 10$ 
- $\left|K\right|$ cardinalità dell'insieme delle chiavi: $\left|K\right| = m$
- $P(k=i)$ distibuzione di probabilità con cui sono generate le chiavi : $P(k=i) = \frac{1}{\left|K\right|}  \quad \forall i \in K$

# 2. Hash Table
Se il numero effettivo di chiavi memorizzate $n$ è inferiore alla cardinalità dell'insieme delle chiavi $\left|K\right|$, le **hash table** risultano più efficienti da un punto di vista di occupazione di memoria rispetto alle tabelle ad indirizzamento diretto, in quanto tipicamente utilizzano un vettore di dimensione $m$ proporzionale ad $n$.

Invece che utilizzare direttamente la chiave come indice dell'array in cui andare a memorizzare un elemento, le tabelle hash utilizzano la chiave per calcolare tale indice, applicando ad essa una **funzione hash**: <br>
<center>$h : K \rightarrow \{0, 1, \dots, m-1\}$</center>

Due chiavi distinte possono essere associate allo stesso indice dell'array. Tale evento è detto **collisione** e deve essere gestito attraverso due possibili approcci: *concatenamento* e *indirizzamento aperto*.

## 2.1. Chained Hash Table
Il metodo del **concatenamento** prevede di memorizzare in una lista concatenata tutti gli elementi la cui chiave genera lo stesso indice, ovvero tutti gli elementi aventi lo stesso valore hash $h(k)$. In questo modo la tabella è costituita da un vettore contenente $m$ puntatori ad $m$ liste. La generica lista $j$ conterrà tutti gli elementi per cui risulta $h(k) = j$.

Supponendo che le liste siano doppiamente concatenate, il tempo di esecuzione delle operazioni di inserimento e cancellazione di un elemento è $O(1)$. Per quanto riguarda la ricerca è necessario distinguere caso medio e caso peggiore:
- Nel caso peggiore la ricerca è invece eseguita in un tempo $O(n)$.
- Nel caso medio il tempo di esecuzione dipende da come la funzione di hash distribuisce mediamente le chiavi da memorizzare tra le diverse liste. Sotto determinate ipotesi il tempo di esecuzione nel caso medio è $O(1)$. 

### Insert
```java
//  https://www.baeldung.com/java-collections-complexity
public boolean insert(Element element) {
    int i = hash(element.getKey());
    table[i].addFirst(element); // O(1)
    n = n + 1;
    return true;
}
```

### Search
```java
public Element search(int key) {
    Element e = new Element(key);
    int k = hash(key);

    int index = table[k].indexOf(e); // O(n)

    if(index != -1) {
        return table[k].get(index); // O(n)
    } else
        return null;
}
```

### Delete
```java
public boolean delete(Element element) {
    int i = hash((int) element.getKey());
    boolean result;
    result = table[i].remove(element);  // O(1)

    if(result) {
        n = n - 1;
        return true;
    } else
        return false;
}
```

## Simple Uniform Hashing
Un'ipotesi *necessaria* affinché il tempo di esecuzione di un'operazione di ricerca nel caso medio sia $O(1)$ è detta **hashing uniforme semplice**. Secondo tale ipotesi, qualsiasi elemento ha la stessa probabilità di essere assegnato ad una qualsiasi delle $m$ liste della tabella. In altre parole, la probabilità che un elemento avente chiave $i$ sia assegnato ad una generica lista $j$ è pari a: 
<center>$P(h(k_{i}) = j) = \frac{1}{m} \quad \forall i \in K, \forall j \in [0, m-1]$</center>


In generale, la sola ipotesi di hashing uniforme semplice non è sufficiente a garantire un tempo di esecuzione $O(1)$ nel caso medio. Sotto tale condizioni il tempo di esecuzione dipende da un parametro $\alpha$ detto **fattore di carico**, definito come il rapporto tra il numero di chiavi effettivamente memorizzate e il numero di liste dell'hash table:
<center>$\alpha = \frac{n}{m}$</center>

In altre parole, $\alpha$ rappresenta il numero medio di elementi contenuti in ogni lista. <br> Nel caso in cui le collisioni sono risolte con il concatenamento e nell'ipotedi di hashing uniforme semplice, il tempo di esecuzione di un'operazione di ricerca nel caso medio è $\Theta(1 + \alpha)$

Evidentemente, per come è stato definito $\alpha$, se $n = O(m)$, si ha $\alpha = O(1)$. Di conseguenza, se il numero di liste della tabella hash è almeno proporzionale al numero di elementi memorizzati la ricerca nel caso medio è eseuita in un tempo costante.

## 2.2. Open Address Hash Table
Il metodo dell'**indirizzamento aperto** prevede di memorizzare tutti gli elementi all'interno del vettore. Le collisioni sono risolte consentendo ad ogni elemento avente una certa chiave di poter essere memorizzato in un insieme di possibili posizioni. 

A tale scopo, si estende la funzione di hash come segue:
<center>$h : K \times \{0, 1, \dots, m-1\} \rightarrow \{0, 1, \dots, m-1\}$</center>

Ad ogni elemento è associata quindi una sequenza di posizioni $<h(k,0), h(k,1), ..., h(k,m-1)>$, detta **sequenza di probing** (ispezione). Per accedere ad un elemento è necessario esaminare in successione le posizioni della sequenza ad esso associata. Tale processo è detto **probing**.

Come nel caso dell'inisizzamento aperto, il numero massimo di elementi che è possibile memorizzare coincide con la dimensione $m$ del vettore. Inoltre, per poter riempire completamente l'array, è necessario che la sequenza di ispezione sia una permutazione dei valori $0, 1, \dots, m-1$.

### Insert
```java
public boolean insert(Element element) {
    int i = 0;
    int j = 0;

    do {
        j = hash(element.getKey(), i);

        if(element.equals(table[j])) {  // garantisce che le chiavi siano distinte
            return false;
        }

        if(table[j] == null || DELETED.equals(table[j])) {
            table[j] = element;
            n = n + 1;
            return true;
        } else {
            i = i + 1;
        }
    } while(i < m);
    return false;
}
```

### Search
```java
public Element search(int key) {
    int i = 0;
    int j = 0;

    do {
        j = hash(key, i);

        if(table[j] != null && table[j].getKey() == key) {
            return table[j];
        } else {
            i = i + 1;
        }
    } while(i < m && table[j] != null);
    return null;
}
```

### Delete
```java
public boolean delete(Element element) {
    int i = 0;
    int j = 0;

    do {
        j = hash(element.getKey(), i);

        if(table[j] != null && table[j].equals(element)) {
            table[j] = DELETED;
            n = n - 1;
            return true;
        } else {
            i = i + 1;
        }
    } while(i < m && table[j] != null);
    return false;
}
```

### 2.2.1. Linear Probing Hash Table
Un primo esempio di indirizzamento aperto è quello del **linear probing** (ispezione lineare), in cui la funzione di hash è:
<center>$h(k, i) = (h_1(k) + i) \mod m \qquad \forall i \in [0, m-1]$</center>
dove $h_1(k)$ è una funzione di hash ausiliaria. <br>
In quest'analisi è stato scelto:
<center>$h_1(k) = k \mod m \qquad \forall k \in K$</center>

I principali limiti dell'ispezione lineare sono i seguenti:
- Si utilizzano solo $m$ delle $m!$ possibili permutazioni. La sequenza di ispezione di un generico elemento avente chiave k risulta infatti: 
<center>$<h_1(k), h_1(k)+1, \dots, m-1, 0, 1, \dots, h_1(k)-1>$</center>
Il valore $h_1(k)$ determina l'intera sequenza di probing, per cui sequenze distinte differiscono unicamente per il punto di partenza.
- Soffre di un problema noto come *addensamento primario*. Può verificarsi una concentrazione di elementi in un certo sottointervallo dell'array.

- $m = 10^6$
- $n \in \left[0, m-10^3\right]$

- *op_iteration* $= 10$ 
- $\left|K\right| = 10^9$
- $P(k=i) = \frac{1}{\left|K\right|}  \quad \forall i \in K$

````python
df_linear_insert = pd.read_csv("HashTable\Open\Linear\Insert.csv", sep = '\t')

df_linear_insert = df_linear_insert.assign(alpha = lambda x: x.n/x.m)
df_linear_insert = df_linear_insert.drop(columns = ['m','n','result'])
df_linear_insert = df_linear_insert.groupby('alpha').median()
df_linear_insert.reset_index(inplace = True)
````

````python
df_linear_search_avg_tf = pd.read_csv("HashTable\Open\Linear\SearchAverage.csv", sep = '\t')

df_linear_search_avg_true = df_linear_search_avg_tf.query('result == True')

df_linear_search_avg_true = df_linear_search_avg_true.assign(alpha = lambda x: x.n/x.m)
df_linear_search_avg_true = df_linear_search_avg_true.drop(columns = ['m','n','result'])
df_linear_search_avg_true = df_linear_search_avg_true.groupby('alpha').median()
df_linear_search_avg_true.reset_index(inplace = True)
df_linear_search_avg_true = df_linear_search_avg_true.assign(result = True)

df_linear_search_avg_false = df_linear_search_avg_tf.query('result == False')

df_linear_search_avg_false = df_linear_search_avg_false.assign(alpha = lambda x: x.n/x.m)
df_linear_search_avg_false = df_linear_search_avg_false.drop(columns = ['m','n','result'])
df_linear_search_avg_false = df_linear_search_avg_false.groupby('alpha').median()
df_linear_search_avg_false.reset_index(inplace = True)
df_linear_search_avg_false = df_linear_search_avg_false.assign(result = False)
````

Tracciando l'andamento del tempo di esecuzione in funzione del fattore di carico $\alpha$ è possibile osservare la complessità asintotica dell'operazione di *search*. In particolare sono stati distinti i casi di ricerca senza successo e ricerca con successo:
- Nel caso di ricerca senza successo (esito *false*), la complessità è $O(\frac{1}{1-\alpha})$.
- Nel caso di ricerca con successo (esito *true*), la complessità è $O(\frac{1}{\alpha} \ln {\frac{1}{1-\alpha}})$.

````python
df_linear_search_avg_tf = df_linear_search_avg_true.append(df_linear_search_avg_false, ignore_index = True)

groups = df_linear_search_avg_tf.groupby('result')
for name, group in groups:
    plt.scatter(group['alpha'], group['time'], s = 15, label=name)
plt.xlabel('alpha')
plt.ylabel('time (ns)')
plt.title('Search Average Case')
plt.legend()
````

### 2.2.2. Quadratic Probing Hash Table
Un altro esempio di indirizzamento aperto è quello del **quadratic probing** (ispezione quadratica), in cui la funzione di hash è:
<center>$h(k, i) = (h_1(k) + c_1 i + c_2 i^2) \mod m \qquad \forall i \in [0, m-1]$</center>
dove $h_1(k)$ è una funzione di hash ausiliaria. <br>
In quest'analisi è stato scelto:
<center>$h_1(k) = k \mod m \qquad \forall k \in K$</center>
<center>$c_1 = 1 , \quad c_2 = 3$</center>

Come nel linear probing, si utilizzano solo $m$ delle $m!$ possibili permutazioni e il valore $h_1(k)$ determina l'intera sequenza di probing.

Tale soluzione soffre inoltre di un problema noto come *addensamento secondario*, meno grave rispetto all'addensamento primario del linear probing.

### 2.2.3. Double Hashing 
Un ultimo esempio di indirizzamento aperto è quello del **double hashing** (doppio hashing), in cui la funzione di hash è:
<center>$h(k, i) = (h_1(k) + i \cdot h_2(k)) \mod m \qquad \forall i \in [0, m-1]$</center>
dove $h_1(k)$ è una funzione di hash ausiliaria. <br>
In quest'analisi è stato scelto:
<center>$h_1(k) = k \mod m \qquad \forall k \in K$</center>
<center>$h_2(k) = 1 + (k \mod m-1) \qquad \forall k \in K$</center>

A differenza delle soluzioni precedenti, si utilizzano $m^2$ delle $m!$ possibili permutazioni. Inoltre, il valore della chiave non determina solo il punto di partenza ma anche l'offset da aggiungere ad ogni iterazione durante l'ispezione.


### Valutazione delle collisioni
L'analisi è stata effettuata considerando i seguenti parametri:
- $m = 10^4$
- $n \in \left[1, m\right]$
- $\left|K\right| = 10^8$
- $P(k=i) = \frac{1}{\left|K\right|}  \quad \forall i \in K$


````python
df_linear_collision = pd.read_csv("HashTable\Open\Linear\InsertCollision.csv", sep = '\t')
df_linear_collision = df_linear_collision.assign(alpha = lambda x: x.n/x.m)

df_quadratic_collision = pd.read_csv("HashTable\Open\Quadratic\InsertCollision.csv", sep = '\t')
df_quadratic_collision = df_quadratic_collision.assign(alpha = lambda x: x.n/x.m)

df_double_collision = pd.read_csv("HashTable\Open\Double\InsertCollision.csv", sep = '\t')
df_double_collision = df_double_collision.assign(alpha = lambda x: x.n/x.m)
````

Di seguito si propone un'analisi sulle prestazioni dei diversi meccanismi di indirizzamento aperto, tracciando il numero di collisioni in funzione del fattore di carico $\alpha$.
